 ### Identificación de Fases en KDD y CRISP-DM

¿Cómo aplicaría el proceso KDD a estos datos?
¿Qué pasos seguirías antes de extraer patrones?
Para aplicar el proceso KDD (Knowledge Discovery in Databases) a estos datos, seguiría los siguientes pasos:
1.	Selección:
Seleccionar los datos relevantes para el análisis, como la edad, el plan contratado, el consumo de datos, los reclamos y el estado de cuenta.
2.	Preprocesamiento:
Limpiar los datos buscando valores faltantes, inconsistentes o duplicados. Además, se deben transformar los datos categóricos en valores numéricos o etiquetas entendibles por algoritmos de análisis.
3.	Transformación:
Preparar los datos para la minería de datos. Esto incluye normalizar valores, crear nuevas variables (como una columna binaria “es_moroso”), o agrupar los datos en rangos.
4.	Minería de Datos:
Aplicar técnicas analíticas para identificar patrones, como analizar si los morosos tienden a tener ciertos planes o niveles de consumo.
5.	Interpretación y Evaluación:
Interpretar los resultados obtenidos y evaluar su utilidad para tomar decisiones. Por ejemplo, detectar si los planes más económicos presentan mayor morosidad.


¿Qué fase de CRISP-DM corresponde a cada acción en este análisis?

Acción:

Comprender qué información contiene la tabla
Verificar valores nulos o inconsistentes
Codificar variables categóricas
Generar histogramas y gráficos para explorar los datos
Analizar si los clientes morosos tienen patrones en común
Concluir si hay relación entre consumo de datos y estado de cuenta

Fase de CRISP-DM:

Comprensión del negocio
Comprensión de los datos
Preparación de los datos
Comprensión de los datos
Modelado / Evaluación
Evaluación / Presentación de resultados


Interpretación de los datos

Luego de analizar los datos, podemos observa que los clientes morosos tienden a tener planes más económicos (Básico o Estándar) y un consumo de datos entre bajo a medio. También, los clientes sin reclamos y con mayor consumo de datos suelen estar al día en sus pagos. Esto lo que nos indica es que los clientes que usan más el servicio y están más satisfechos, cumplen con sus pagos regularmente.
Fase más relevante en KDD:
Minería de datos: En mi opinión esta fase es la más relevante, ya que permite identificar patrones y relaciones, como la relación entre el tipo de plan y el estado de cuenta. Estos conocimientos que se extraen ayudan a tomar decisiones importantes.

Fase más relevante de CRISP-DM:
Comprensión de los datos:
Con esta fase es más fácil entender la calidad de los datos, identificar errores y preparar mejor el análisis posterior.


A continuación los codigos y gráficos obtenidos:

```python
import pandas as pd

# copiar manualmente el DataFrame
data = {
    'Cliente': [1, 2, 3, 4, 5],
    'Edad': [25, 40, 32, 22, 35],
    'Plan Contratado': ['Básico', 'Premium', 'Estándar', 'Básico', 'Premium'],
    'Consumo de Datos (GB)': [5, 50, 10, 7, 45],
    'Reclamos en el Último Año': [1, 0, 2, 1, 0],
    'Estado de Cuenta': ['Pagado', 'Pagado', 'Moroso', 'Moroso', 'Pagado']
}

df = pd.DataFrame(data)

# Muestra el Dataframe
df
```

```python
# Verificar valores faltantes
print("Valores faltantes por columna:")
print(df.isnull().sum())

# Verificar duplicados
print("\n¿Existen filas duplicadas?")
print(df.duplicated().any())
```

```python
# Codificar variables categóricas con Label Encoding (opcional: usar OneHotEncoding si es para modelado)
df['Plan Codificado'] = df['Plan Contratado'].map({'Básico': 0, 'Estándar': 1, 'Premium': 2})
df['Estado Codificado'] = df['Estado de Cuenta'].map({'Pagado': 0, 'Moroso': 1})

df
```

```python
import matplotlib.pyplot as plt

# Histograma del consumo de datos
plt.hist(df['Consumo de Datos (GB)'], bins=5, color='skyblue', edgecolor='black')
plt.title('Histograma de Consumo de Datos (GB)')
plt.xlabel('Consumo de Datos (GB)')
plt.ylabel('Cantidad de Clientes')
plt.grid(True)
plt.show()
```

```python
# Conteo de clientes morosos vs pagados
estado_counts = df['Estado de Cuenta'].value_counts()

# Gráfico de barras
estado_counts.plot(kind='bar', color=['green', 'red'])
plt.title('Estado de Cuenta de los Clientes')
plt.xlabel('Estado de Cuenta')
plt.ylabel('Cantidad de Clientes')
plt.grid(axis='y')
plt.show()
```

```python
# Agrupación por Estado de Cuenta para analizar características de los morosos
morosos = df[df['Estado de Cuenta'] == 'Moroso']
morosos_stats = morosos.describe(include='all')

print("Clientes morosos - características:")
print(morosos_stats)

# Relación entre consumo de datos y estado de cuenta
import seaborn as sns

sns.boxplot(x='Estado de Cuenta', y='Consumo de Datos (GB)', data=df)
plt.title('Consumo de Datos según Estado de Cuenta')
plt.show()
```
